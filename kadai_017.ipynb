```python
class Human:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(self.name,"は大人です。")
        else:
            print(self.name,"は大人ではありません。")

humans = [
    Human("山田",43),
    Human("鈴木",28),
    Human("加藤",31),
    Human("野口",17)
]

for person in humans:
    person.check_adult()
```
